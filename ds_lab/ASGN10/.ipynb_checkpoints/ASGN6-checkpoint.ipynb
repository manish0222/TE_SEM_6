```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from scipy.stats import skew
```

```python
# Check the available inbuilt datasets
print(sns.get_dataset_names())
df=sns.load_dataset("iris")
df.head()
```

```python
#List down the features and their types (e.g., numeric, nominal) available in thedataset.
df.dtypes
```

```python
df.isna().sum()
```

```python
# Get numeric columns (excluding the last column)
numeric_columns = df.select_dtypes(include=np.number).columns.tolist()

# Plot histogram and calculate skewness for each column
for column in numeric_columns:
    plt.figure(figsize=(8, 4))
    sns.histplot(df[column], kde=True, color='skyblue')
    plt.title(f'Histogram of {column}')
    plt.xlabel(column)
    plt.ylabel('Frequency')
    plt.show()
    
    skewness = skew(df[column])
    print(f"Skewness of {column}: {skewness}")
    
    if skewness > 0:
        print(f"{column} is positively skewed.")
    elif skewness < 0:
        print(f"{column} is negatively skewed.")
    else:
        print(f"{column} is normally distributed.")
```

```python
# Create box plots for each feature
plt.figure(figsize=(10, 8))
sns.boxplot(data=df)
plt.show()
```

```python
# Check for null values
print(df.isnull().sum())

# If there are null values, you can handle them by dropping or imputing
# Get numeric columns (excluding the last column)
numeric_columns = df.select_dtypes(include=np.number).columns.tolist()

# Fill null values with the mean of each numeric column
df.fillna(df[numeric_columns].mean(), inplace=True)
df
```
