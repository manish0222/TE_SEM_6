# Data Analytics I
1. Create a Linear Regression Model using Python/R to predict home prices using Boston Housing
Dataset (https://www.kaggle.com/c/boston-housing). 
2. The Boston Housing dataset contains
information about various houses in Boston through different parameters. There are 506 samples
and 14 feature variables in this dataset.
3. The objective is to predict the value of prices of the house using the given features.

```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.metrics import r2_score,mean_squared_error
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
```

```python
df=pd.read_csv("https://raw.githubusercontent.com/manish0222/TE_SEM_6/main/ds_lab/ASGN7/BostonHousing.csv")
df
```

```python
sns.heatmap(df.corr(),annot=True)
```

```python
df.describe()
```

```python
df.info()
```

```python
df.dtypes
```

```python
# x=df[['age']] #simple linear regression
x=df.drop(['medv'],axis=1)
y=df['medv']
x_train,x_test,y_train,y_test=train_test_split(x,y,random_state=0,test_size=0.2)
#80% training
```

```python
model=LinearRegression()
model.fit(x_train,y_train)
```

```python
y_predict=model.predict(x_train)
```

```python
mse=mean_squared_error(y_train,y_predict)
rse=r2_score(y_train,y_predict)
print(f"mse is {mse} r2score is {rse}")
#low mse=better
#high r2 better score
```

```python
plt.figure(figsize=(8,6))
plt.title("Regression line")
plt.scatter(y_train,y_predict,color='green',label='points')
plt.plot([min(y_train),max(y_train)],[min(y_train),max(y_train)],color='red',label='Line')
plt.xlabel("Actual")
plt.ylabel("Predicted")
plt.legend()
plt.show()
```
