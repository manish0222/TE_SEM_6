## Data Visualization I
1. Use the inbuilt dataset 'titanic'. The dataset contains 891 rows and contains information
about the passengers who boarded the unfortunate Titanic ship. 
2. Use the Seaborn library
to see if we can find any patterns in the data.
3. Write a code to check how the price of the ticket (column name: 'fare') for each
passenger is distributed by plotting a histogram.

```python
import numpy as np
import pandas as pd
import seaborn as sns
from scipy import stats
```

```python
df=sns.load_dataset("titanic")
df
```

```python
df.describe()
```

```python
df.head(5)
```

```python
df.isna().sum()
```

```python
df.dtypes
```

```python
numeric_column=df.select_dtypes(include=np.number).columns.tolist()
print("Numeric Columns are ",numeric_column)
df[numeric_column]=df[numeric_column].fillna(df[numeric_column].mean())
df.isna().sum()
## filled numeric columns with the null values with it's columns mean
```

```python
df=df.ffill()
df=df.bfill()
df.isna().sum()
## removed NaN for object datatypes
```

```python
sns.histplot(data=df, x="sex",bins=0.5)
#analysis no of males in ship > no of females
```

```python
df.plot(kind='hist',x='pclass',y='fare')
# maximum Ticket purchased had prices in 0-100 range
```

```python
df.plot(kind='hist',y='pclass')
## more pclass 3 than 2 than 1
```

```python
sns.boxplot(x='pclass', y='fare', data=df)
## more people of pclass one paid more price than 2 than 3
```

```python
# sns.barplot(x='sex', y='survived', data=df)
sns.barplot(x='sex', y='age',hue='survived',data=df)
#large number of females survived
```

```python
sns.boxplot(x='pclass', y='age', hue='survived', data=df,gap=0.25)
#more people of class 1 survived
```

```python
sns.scatterplot(x='age', y='fare', data=df)
## max price was for 20-50 age group people
```

```python
# Observation: Most passengers which survived or didn't survive embarked the Titanic from Southampton

# Embarked vs. Survived
sns.countplot( data=df , x="embark_town" , hue="survived" )
```

```python
sns.displot(df['sex'])  
```

```python
# Create line plot
sns.lineplot(data=df,x='age',y='fare')
```

```python
sns.violinplot(data=df,x='pclass',y='fare')
```

```python
sns.jointplot(data=df,x='pclass',y='fare',hue='survived')
```
