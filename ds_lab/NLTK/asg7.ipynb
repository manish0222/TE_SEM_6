```python
nltk.download('punkt')
nltk.download('stopwords')
nltk.download('averaged_perceptron_tagger')
nltk.download('wordnet')
```

```python
with open('doc_01.txt','r') as file:
    doc=file.read()

with open('doc_02.txt','r') as file:
    doc2=file.read()
```

```python
doc=re.sub('[\W_]',' ',doc)
doc2=re.sub('[\W_]',' ',doc2)
```

```python
word_token=nltk.word_tokenize(doc)
word_token
```

```python
#sentence tokenization
sent_token=nltk.sent_tokenize(doc)
sent_token
```

```python
stopwords=set(nltk.corpus.stopwords.words('english'))
filtered_words=[]
for token in word_token:
    if token not in stopwords:
        filtered_words.append(token)

filtered_words
```

```python
pos_tagging=nltk.pos_tag(filtered_words)
pos_tagging
```

```python
stemmer=nltk.stem.PorterStemmer()
for w in filtered_words:
    print(w," :",stemmer.stem(w))
```

```python
lemmatizer=nltk.stem.WordNetLemmatizer()
for w in filtered_words:
    print(w," :",lemmatizer.lemmatize(w))
```

```python
def calculate_tf(doc):
    word_token=nltk.word_tokenize(doc)
    word_freq=dict()
    for w in word_token:
        word_freq[w]=word_token.count(w)
    tf=dict()
    for w,c in word_freq.items():
        tf[w]=c/len(word_freq)
    return tf    
    
```

```python
tf1=calculate_tf(doc)
tf1
```

```python
tf2=calculate_tf(doc2)
tf2
```

```python
def calculate_idf(docList):
    all_tokens=[]
    for d in docList:
        all_tokens+=nltk.word_tokenize(d)
    nodocs=len(docList)
    idf=dict()
    for t in all_tokens:
        f=0
        for d in docList:
            if t in nltk.word_tokenize(d):
                f+=1
        idf[t]=math.log(nodocs/f)
    return idf
```

```python
idf=calculate_idf([doc,doc2])
idf
```
