### Filling Null values and 2 methods

```python
columns_to_fill = ['Car', 'BuildingArea', 'YearBuilt' ]
df[columns_to_fill] = df[columns_to_fill].fillna(df[columns_to_fill].mean())
print("The shape is ",df.shape)
df
```

```python
print("The shape is ",df.shape)
df.isna().sum()
df=df.ffill()  #forward fill the null values by previous non null values
df=df.bfill()  # backward fill in background
df
```

# Categorical Replacement

```python
#Replace Method  
#                                    NOT used Dummy Method
df['Rooms'].replace([1,2,3,4,5,6,7,8,10],
                    ['1BHK','2BHK','3BHK','4BHK','5BHK','6BHK','7BHK','8BHK','10BHK'],inplace=True)
df
```

```python
!pip install category_encoders

import category_encoders as ce

# Assume df is your DataFrame with 'Rooms' column modified
# Display the original DataFrame
print("Original DataFrame:")
df
```

```python
# 1. Label Encoding for 'Rooms'
label_encoder_rooms = LabelEncoder()
df['Rooms_LabelEncoded'] = label_encoder_rooms.fit_transform(df['Rooms'])
new_arr=['Rooms_LabelEncoded','Rooms']
# Display the DataFrame after Label Encoding for 'Rooms'
print("\nDataFrame after Label Encoding for 'Rooms':")
df[new_arr]
```

```python
# 2. One-Hot Encoding for 'Rooms'
df_one_hot_rooms = pd.get_dummies(df, columns=['Rooms'], prefix=['Rooms'])

# Display the DataFrame after One-Hot Encoding for 'Rooms' 
# number of colums increases now of categories = number of new columns
print("\nDataFrame after One-Hot Encoding for 'Rooms':")
df_one_hot_rooms
```

```python
# 3. Ordinal Encoding for 'Rooms'
ordinal_mapping_rooms = {'1BHK': 0, '2BHK': 1, '3BHK': 2, '4BHK': 3, '5BHK': 4, '6BHK': 5, '7BHK': 6, '8BHK': 7, '10BHK': 8}
df['Rooms_OrdinalEncoded'] = df['Rooms'].map(ordinal_mapping_rooms)
new_arr.append('Rooms_OrdinalEncoded')
# Display the DataFrame after Ordinal Encoding for 'Rooms'
print("\nDataFrame after Ordinal Encoding for 'Rooms':")
df[new_arr].head(25)
```

```python
# 4. Binary Encoding for 'Rooms'
binary_encoder_rooms = ce.BinaryEncoder(cols=['Rooms'])
df_binary_rooms = binary_encoder_rooms.fit_transform(df)

# Display the DataFrame after Binary Encoding for 'Rooms'
print("\nDataFrame after Binary Encoding for 'Rooms':")
df_binary_rooms
```

```python
# Create the 'Copy' column and set its values to be the same as 'Distance'
df['Distance_copy'] = df['Distance']
df
```

```python
# normalization is a process to make values in 0 and 1 range so its effective for data interpretation
# Example: Normalize the 'BP_Normal' column
df['Distance_copy'] = (df['Distance_copy'] - df['Distance_copy'].min()) / (df['Distance_copy'].max() - df['Distance_copy'].min())
# Display the DataFrame with the normalized 'BP_Normal' column
df
```

```python
# Drop the 'Distance_copy' column with inplace
#axis =1 column    0 is row

df.drop(['Distance_copy'], axis=1, inplace=True)
df
```
