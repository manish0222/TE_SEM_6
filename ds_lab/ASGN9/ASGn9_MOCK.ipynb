```python
# hBBSCD ATCOunt
# line violin joint displot
import numpy as np
import pandas as pd
import seaborn as sns
```

```python
df=sns.load_dataset('titanic')
df.head()
```

```python
df.isna().sum()
```

```python
numric_col=df.select_dtypes(include=np.number).columns.tolist()
numric_col
```

```python
df.dtypes
```

```python
df[numric_col]=df[numric_col].fillna(df[numric_col].mean())
df.isna().sum()
```

```python
print("shape ",df.shape)
df=df.ffill()
df=df.bfill()
df.isna().sum()
```

```python
df.head()
```

```python
sns.histplot(data=df,x='sex',hue='survived',bins=0.5)
```

```python
sns.barplot(data=df,x='sex',y='pclass',hue='survived')
```

```python
sns.boxplot(data=df,x='sex',y='age',hue='survived')
```

```python
sns.barplot(data=df,x='survived',y='age',hue='pclass')
```

```python
sns.scatterplot(data=df,x='age',y='fare')
```

```python
sns.displot(data=df,x='pclass')
```

```python
sns.violinplot(data=df,x='sex',y='pclass',hue='survived')
```

```python
sns.jointplot(data=df,x='sex',y='pclass')
```

```python
sns.lineplot(data=df,x='embark_town',y='alive')
```
